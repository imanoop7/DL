```python
path="game_of_thrones.txt"
```

```python
data = open(path).read()
```

```python
corpus = data.lower().split("\n")
```

```python
size=int(len(corpus)/10)
```

```python
docs=corpus[0:size]
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.word_index
```

```python
input_sequences=[]
for sentence in docs:
  tokenized_sentence= tokenizer.texts_to_sequences([sentence])[0]
  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
total_words=len(tokenizer.word_index)+1
```

```python
total_words
```

```python
max_len=max([len(x) for x in input_sequences])
```

```python
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

padded_input_sequence = pad_sequences(input_sequences,max_len,padding='pre')
```

```python
X= padded_input_sequence[:,:-1]
Y=padded_input_sequence[:,-1]
```

```python
X.shape
```

```python
Y
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y=to_categorical(Y,num_classes=total_words)
```

```python
y[0]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
model= Sequential()
```

```python
model.add(Embedding(total_words,100,input_length=max_len-1))
model.add(LSTM(150))
model.add(Dense(total_words,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=20,validation_split=0.2)
```

```python
import numpy as np
```

```python
text= "mail me"
next_word =50
for _ in range(next_word):
  tokenized_text= tokenizer.texts_to_sequences([text])[0]
  padded_text= pad_sequences([tokenized_text],maxlen=max_len-1,padding='pre')
  prediction = np.argmax(model.predict(padded_text))

  for word, index in tokenizer.word_index.items():
    if index == prediction:
      text =  text + " " + word
      print(text)
```
